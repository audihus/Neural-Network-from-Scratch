```python
#NEURAL NETWORKS

def init_params(hidden_neuron=32):
    W1 = np.random.rand(hidden_neuron,12) * 0.01 #inisialisasi weight untuk layer 1
    b1 = np.random.rand(hidden_neuron,1) #inisialisasi bias untuk layer 1

    W2 = np.random.rand(1,hidden_neuron) * 0.01 #inisialisasi weight untuk layer 2
    b2 = np.random.rand(1,1) #inisialisasi bias untuk layer 2

    return W1, b1, W2, b2

def leaky_relu(Z):
    return np.maximum(0.1*Z, Z)

def linear(Z):
    return Z

def deriv_leaky_relu(Z, alpha=0.01):
    grad = np.ones_like(Z)
    grad[Z < 0] = alpha
    return grad

def relu(Z):
    return np.maximum(0, Z)

def deriv_relu(Z):
    return Z > 0

def forward(W1, b1, W2, b2, X):
    '''
    Layer 1
    '''
    #Z1 adalah sum dari layer 1
    Z1 = W1.dot(X) + b1
    #A1 adalah output dari layer 1
    A1 = leaky_relu(Z1)

    '''
    Layer 2
    '''
    Z2 = W2.dot(A1) + b2
    A2 = linear(Z2)

    return Z1, A1, Z2, A2

def backpropagation(Z1, A1, Z2, A2, W2, X, Y):
    m = Y.size
    dZ2 = A2 - Y
    dW2 = (1/m) * dZ2.dot(A1.T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True) #kenapa gini


    dZ1 = W2.T.dot(dZ2) * deriv_leaky_relu(Z1)
    dW1 = (1/m) * dZ1.dot(X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True) #kenapa gini

    return dW1, db1, dW2, db2

def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    new_W1 = W1 - alpha * dW1
    new_b1 = b1 - alpha * db1
    new_W2 = W2 - alpha * dW2
    new_b2 = b2 - alpha * db2
    
    return new_W1, new_b1, new_W2, new_b2

```

```python
#GRADIENT DESCENT
# fungsi untuk mengubah output aktivasi menjadi prediksi akhir
def get_prediction(A2):
    """
    Karena ini regresi, prediksi = output langsung
    """
    return A2

# fungsi untuk menghitung error (contoh MSE)
def get_error(prediction, Y, loss="mse"):
    """
    Menghitung loss antara prediksi dan target.
    prediction, Y: shape (1, m)
    """
    if loss == "mse":
        return np.mean((prediction - Y) ** 2)
    elif loss == "mae":
        return np.mean(np.abs(prediction - Y))
    else:
        raise ValueError("Unknown loss type")



def batch_gradient_descent(X,Y, epochs, alpha,hidden_neuron):
    W1, b1, W2, b2 =  init_params(hidden_neuron)
    for i in range(1, epochs+1):
        Z1, A1, Z2, A2 = forward(W1,b1,W2,b2,X)
        dW1, db1, dW2, db2 = backpropagation(Z1,A1,Z2,A2,W2,X,Y)
        W1, b1, W2, b2 = update_parameters(W1,b1,W2,b2,dW1,db1,dW2,db2,alpha)

        if i % 10 ==0:
            pred = get_prediction(A2)
            mse = get_error(pred, Y, loss="mse")
            mae = get_error(pred, Y, loss="mae")
            print(f"Iter {i:4d} | MSE = {mse:.4f} | MAE = {mae:.4f}")
    
    return W1, b1, W2, b2


def stoch_gradient_descent(X, Y, epochs, alpha, hidden_neuron, shuffle=True, seed=42):
    """
    Stochastic / Mini-batch Gradient Descent
    - batch_size=1    -> SGD murni
    - batch_size>1    -> mini-batch SGD
    """

    rng = np.random.default_rng(seed)

    n = X.shape[1]  # jumlah record
    W1, b1, W2, b2 = init_params(hidden_neuron)

    for i in range(1, epochs+1):
        idx = np.arange(n) #generate indeks data 1,2,3,4,...
        if shuffle:
            rng.shuffle(idx) #indeks di-shuffle 59, 21, 56, 18,...

        # --- iterasi per mini-batch ---
        for start in range(0, n, 1):
            batch_idx = idx[start:start + 1] #ngambil 50 index 
            Xb = X[:, batch_idx]
            Yb = Y[:, batch_idx] if Y.ndim == 2 else Y[batch_idx]  # sesuaikan bentuk

            Z1, A1, Z2, A2 = forward(W1, b1, W2, b2, Xb)
            dW1, db1, dW2, db2 = backpropagation(Z1, A1, Z2, A2, W2, Xb, Yb)
            W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)

        # --- monitoring loss/metric per i (pakai full data) ---
        if (i % 10) == 0:
            _, _, _, A2_full = forward(W1, b1, W2, b2, X)
            pred = get_prediction(A2_full)
            mse = get_error(pred, Y, loss="mse")
            mae = get_error(pred, Y, loss="mae")
            print(f"Epoch {i:4d} | MSE = {mse:.4f} | MAE = {mae:.4f}")

    return W1, b1, W2, b2

def mini_batch_gradient_descent(X, Y, epochs, alpha, hidden_neuron, batch_size=1, shuffle=True, seed=42):
    """
    Stochastic / Mini-batch Gradient Descent
    - batch_size=1    -> SGD murni
    - batch_size>1    -> mini-batch SGD
    """

    rng = np.random.default_rng(seed)

    n = X.shape[1]  # jumlah record
    W1, b1, W2, b2 = init_params()

    for i in range(1, epochs+1):
        idx = np.arange(n) #generate indeks data 1,2,3,4,...
        if shuffle:
            rng.shuffle(idx) #indeks di-shuffle 59, 21, 56, 18,...

        # --- iterasi per mini-batch ---
        for start in range(0, n, batch_size):
            batch_idx = idx[start:start + batch_size] #ngambil 50 index 
            Xb = X[:, batch_idx]
            Yb = Y[:, batch_idx] if Y.ndim == 2 else Y[batch_idx]  # sesuaikan bentuk

            Z1, A1, Z2, A2 = forward(W1, b1, W2, b2, Xb)
            dW1, db1, dW2, db2 = backpropagation(Z1, A1, Z2, A2, W2, Xb, Yb)
            W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)

        # --- monitoring loss/metric per i (pakai full data) ---
        if (i % 10) == 0:
            _, _, _, A2_full = forward(W1, b1, W2, b2, X)
            pred = get_prediction(A2_full)
            mse = get_error(pred, Y, loss="mse")
            mae = get_error(pred, Y, loss="mae")
            print(f"Epoch {i:4d} | MSE = {mse:.4f} | MAE = {mae:.4f}")

    return W1, b1, W2, b2
```
